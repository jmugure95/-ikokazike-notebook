```python
user_name=kazi.username.str.replace("b'","").str.replace("'","")
user_name.head()
```

```python
users_name=user_name.str.capitalize()
users_name[0:30]
```

```python
kazi.username.mode()
# Username with the highest postings
```

```python
frequented = users_name.value_counts()
frequented.head()
```

```python
frequented[0:20]
```

```python
ax = frequented.head().plot(kind='bar',
                                    figsize=(16,8),
                                    title="TOP 5 USER POSTINGS")
ax.set_xlabel("Username")
ax.set_ylabel("Number of Postings")
plt.show()
```
